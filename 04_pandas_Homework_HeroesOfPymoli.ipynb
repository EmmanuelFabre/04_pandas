
## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
####Purchasing Analysis (Gender)####
#Avg purch count by gender
df_clean = df_orig[["Gender", "Price", "SN", "Age"]]
gender_count = df_clean["Gender"].value_counts()
#gender_count  (debug statement)
```

```python
df_gender_group = df_clean.groupby(["Gender"])
gender_purch_sum = df_gender_group["Price"].sum()
gender_purch_sum
#################
#converting groupby object into DF (sum price by gender)
gender_purch_sum_df = pd.DataFrame(gender_group_sum)
gender_purch_sum_df = gender_purch_sum_df.rename(columns={"Price":"Total Purchase Value"})
gender_purch_sum_df["Total Purchase Value"] = gender_purch_sum_df["Total Purchase Value"].map("${:.2f}".format)

#gender_purch_sum_df #(debug statement)
```

```python
#Final DF for purchasing analysis (gender)
df_final_gender = df_gender_group.mean()     
df_final_gender["Purchase Count"] = gender_count 
df_final_gender = df_final_gender.rename(columns={"Price":"Average Purchase Price"})
df_final_gender["Average Purchase Price"] = df_final_gender["Average Purchase Price"].map("${:.2f}".format)

df_final_gender["Total Purchase Value"] = gender_purch_sum
df_final_gender["Total Purchase Value"] = df_final_gender["Total Purchase Value"].map("${:.2f}".format)

df_final_gender["Avg Total Purchase per Person"] = gender_purch_sum / df_totals
df_final_gender["Avg Total Purchase per Person"] = df_final_gender["Avg Total Purchase per Person"].map("${:.2f}".format)
del df_final_gender["Age"]
df_final_gender
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#####Top Spenders####
spenders_df = pd.DataFrame([])
spenders_df["SN"] = df_orig["SN"] 
spenders_df["Price"] = df_orig["Price"] 
#spenders_df (debug statement)
```

```python
#Purchase sums by user (object)
spenders_group = spenders_df.groupby(["SN"])
spenders_group = spenders_group["Price"].sum()
#spenders_group #(debug statement)
```

```python
#Final Top Spenders DF
spenders_final_df = pd.DataFrame([])
spenders_final_df["Purchase Count"] = spenders_df["SN"].value_counts()

spenders_final_df["Average Purchase Price"] = spenders_group / spenders_df["SN"].value_counts()
spenders_final_df["Average Purchase Price"] = spenders_final_df["Average Purchase Price"].map("${:.2f}".format)

spenders_final_df["Total Purchase Value"] = spenders_group
spenders_final_df["Total Purchase Value"] = spenders_final_df["Total Purchase Value"].astype('float')
spenders_final_df = spenders_final_df.sort_values(["Total Purchase Value"], ascending=False)
spenders_final_df["Total Purchase Value"] = spenders_final_df["Total Purchase Value"].map("${:.2f}".format)  

spenders_final_df.head()
#spenders_final_df.dtypes (debug statement)
```
